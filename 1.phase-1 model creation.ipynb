```python
#pandas is a library and it contains the function related to the database.
import pandas as pd
```

```python
#reading data from the csv file and storing it in 'dataset'
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# 'dataset' is a pandas Dataframe that holds all the ata from salary_data.csv.
dataset
```

```python
#selecting the'years of experience' column as the independent variable (x)
independent=dataset[["YearsExperience"]]
```

```python
#independent holds the input feature(x)
independent
```

```python
#selectinghe dependent variable (output)'salary' column s a Dataframe
dependent=dataset[["Salary"]]
```

```python
#dependent holds the output feature(y)
dependent
```

```python
#import the train_test_split function from sklearn
from sklearn.model_selection import train_test_split
#split the dataset into training ands testing sets and test_size=0.30 means 30% of the data will be used for testing and 70%of training
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#import the LinearRegression class from sklearn
from sklearn.linear_model import LinearRegression
# create an instance of the LinearRegression(model)
regressor=LinearRegression()
#train(fit) the model using the training data
regressor.fit(x_train,y_train)
```

```python
#get the coefficient (weights) of the trained Linear Regression model
weight=regressor.coef_
```

```python
#weight stores the coefficient of the model
weight
```

```python
#get the intercept (bais) of the trained Linear Regression model
bais=regressor.intercept_
```

```python
#bais stores the intercept of the model
bais
```

```python
#predict the target values for the test set using the trained model
y_pred=regressor.predict(x_test)
```

```python
#import r2_score from sk learn to evaluate model performance
from sklearn.metrics import r2_score
#r_score indicates how well the model fits the data 
r_score=r2_score(y_test,y_pred)
```

```python
#print the r_score value to evaluate model performance.
r_score
```

```python
#import the pickle model to save and load python
import pickle
```

```python
#define the file name to save the trained model
filename="finalized_model_linear.sav"
```

```python
#save the trained regressor model to a file in binary format using pickle
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the saved model from the file in binary read mode
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
```

```python
#use the loaded model to predict the output for the input value 15
result=loaded_model.predict([[15]])
```
